```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#coding=utf-8
# data=pd.read_csv("E:\\文档\\/百度网盘\\DataAnalyst.csv")
# 1.加载数据
data = pd.read_csv("E:/文档/百度网盘/DataAnalyst.csv", encoding='GB18030')
# head和tail默认查看前、后5行数据
# print(data.head())
# print(data.tail(6))
print("----------values----------")
# values返回数据帧的numpy形式
# print(data.values)
# print("----------keys----------")
# keys查看列索引信息
# print(data.keys())
print("----------info----------")
# info查看数据维度、列索引名称和类型、数据类型、所占内存空间、以及哪些字段有空的情况
# print(data.info())
# print("----------dtypes----------")
# dtypes查看列索引名称及类型
# print(data.dtypes)

# 2.数据清洗和整理
# unique查重  查看职位id的唯一值
print("----------unique----------")
print(len(data['positionId'].unique()))
print("----------drop_duplicates----------")
# drop_duplicates去重 功能：删除特定列下重复行 参数：subset:默认所有列，keep:first/last 保留第一次或最后一次重复行
# keep: False删除所有重复项 inplace:默认False 是保留一个副本还是在原数据上修改
data_duplicates = data.drop_duplicates(subset='positionId', keep='first')
print(data_duplicates.info())
print(data_duplicates['salary'].head(20))


def cut_word(x, method):
    position = x.find('-')
    if (position != -1):
        bottomSalary = x[:position - 1]
        topSalary = x[position + 1:len(x) - 1]
    else:
        position = x.upper().find('K')
        bottomSalary = x[:position]
        topSalary = bottomSalary
    if (method == 'top'):
        return topSalary
    else:
        return bottomSalary

# bottomSalary = data_duplicates.salary.apply(cut_word)
# 获取最低工资： 工资形式有12k-16k 以及12k以上
# 添加bottomSalary和topSalary列
data_duplicates['bottomSalary'] = data_duplicates.salary.apply(cut_word, method='')
data_duplicates['topSalary'] = data_duplicates.salary.apply(cut_word, method='top')
# 薪资转换成int32
data_duplicates.bottomSalary = data_duplicates['bottomSalary'].astype(int)
data_duplicates.topSalary = data_duplicates['topSalary'].astype(int)
# 添加avgSalary列 axis是apply中的参数，axis=1表示将函数用在行，axis=0则是列
data_duplicates['avgSalary'] = data_duplicates.apply(
    lambda x: (x.bottomSalary + x.topSalary) / 2, axis=1)
print(data_duplicates.info())

# 3.切选出几个我们想要的columns（可根据自己需求切选）
data_clean = data_duplicates[['city', 'companyShortName',
                                  'companySize','education',
                                   'positionName', 'positionLables',
                                   'workYear', 'avgSalary']]
# 4.数据分析和可视化
# print(data_clean.info())
# print(data_clean.city.value_counts())
# print(data_clean.describe())

# 5.数据可视化
# 直方图
%matplotlib inline
# plt.style.use('ggplot')
data_clean.avgSalary.hist(bins=15)
# 箱线图
data_clean.boxplot(column='avgSalary',by='city',figsize=(9,7))
```
